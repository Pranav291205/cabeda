```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import gdown
```

```python
fileid = "1ZaK-OQQOfQLD6v2n2r4PsUGu6YCuYOO7"
gdown.download(f"https://drive.google.com/uc?id={fileid}", "ride_bookings.csv", quiet=False)
df = pd.read_csv("ride_bookings.csv")
print(type(df))
```

```python
nodriveer=df['Booking Status']=='No Driver Found'
totnodriveer=nodriveer.sum()
print(totnodriveer)
```

```python
indices_to_drop = df[df['Booking Status'] == 'No Driver Found'].index
df.drop(indices_to_drop,inplace=True)
```

```python
df.describe(include="all")
```

```python
df['Avg CTAT'].median()
```

```python

df['Avg CTAT'].hist(bins=30)
plt.axvline(df['Avg CTAT'].mean(), color='red', label='Mean')
plt.axvline(df['Avg CTAT'].median(), color='green', label='Median')
plt.legend()
plt.show()
```

```python
df['Avg CTAT'] = df['Avg CTAT'].fillna(df['Avg CTAT'].median())
df.describe()
```

```python
df['Avg VTAT'].median()
df['Avg VTAT']=df['Avg VTAT'].fillna(df['Avg VTAT'].mean())
df.describe()
```

```python
df['Booking Value'].median()
df['Booking Value']=df['Booking Value'].fillna(df['Booking Value'].mean())
df.describe()
```

```python
df.isnull().sum()
```

```python

df['Avg VTAT'].hist(bins=30)
df.boxplot(column='Avg VTAT')
```

```python
completedrides=df['Booking Status']=='Completed'
totcompletedrides=completedrides.sum()
print(totcompletedrides)
labels = ['Completed rides','No Driver Found','Cancelled rides by customer','Cancelled rides by driver','Incomplete rides']
values = [totcompletedrides,totnodriveer,df["Cancelled Rides by Customer"].dropna().sum(),df["Cancelled Rides by Driver"].dropna().sum(),df["Incomplete Rides"].dropna().sum()]
colors = ['lightgreen','lightblue','red','lightyellow','pink']
plt.figure(figsize=(3, 3))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Ride outcome breakdown')
plt.axis('equal')
plt.show()
```

```python
numcols = df.select_dtypes(include=np.number).columns
print(numcols)
```

```python
print("Numeric univariate analysis")
for col in numcols:
    plt.figure(figsize=(6,4))
    df[col].hist(bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
```

```python
df['Vehicle Type'].value_counts()
```

```python
catcols=[col for col in df.select_dtypes(exclude=np.number).columns if df[col].nunique() < 10]
print(catcols)
```

```python
print("Categorial univariate analysis")
for col in catcols:
    plt.figure(figsize=(6,4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df['hour'] = df['Time'].dt.hour
rides_by_hour = df['hour'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=rides_by_hour.index, y=rides_by_hour.values)
plt.title('Trip Frequency by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Rides')
plt.xticks(range(0, 24))
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['day_of_week'] = df['Date'].dt.day_name()
rides_by_day = df['day_of_week'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
sns.barplot(x=rides_by_day.index, y=rides_by_day.values)
plt.title('Trip Frequency by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Rides')
plt.show()
```

```python
fare_hour = df.groupby('hour')['Booking Value'].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(x=fare_hour.index, y=fare_hour.values, marker='o')
plt.title('Average Fare by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Fare')
plt.xticks(range(0, 24))
plt.show()
```

```python
fare_day = df.groupby('day_of_week')['Booking Value'].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=fare_day.index, y=fare_day.values)
plt.title('Average Fare by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Vehicle Type',hue="Payment Method")
plt.title('Trip Frequency by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Rides')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Ride Distance',y='Booking Value',alpha=0.2)
plt.title("Ride distance vs Fare charges")
plt.xlabel("Distance (km)")
plt.ylabel("Fare (₹)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='hour', hue='Vehicle Type')
plt.title('Number of Rides by Hour of Day and Vehicle Type')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Rides')
plt.xticks(range(0, 24))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='day_of_week', hue='Vehicle Type')
plt.title('Number of Rides by Day of Week and Vehicle Type')
plt.xlabel('Day of Week')
plt.ylabel('Number of Rides')
plt.show()
```
